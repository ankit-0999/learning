```python
def recommend_course(title, numrec=10):
    # Clean course titles
    df['Clean_title'] = df['course_title'].str.lower().str.strip()
    title_cleaned = title.lower().strip()

    # Build index mapping for exact match
    course_index = pd.Series(df.index, index=df['Clean_title']).drop_duplicates()

    # Handle exact match
    if title_cleaned in course_index:
        index = course_index[title_cleaned]
        scores = list(enumerate(cosine_sim_mat[index]))
        sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
        selected_course_index = [i[0] for i in sorted_scores[1:numrec+1]]
        selected_course_score = [i[1] for i in sorted_scores[1:numrec+1]]
        rec_df = df.iloc[selected_course_index].copy()
        rec_df['Similarity_Score'] = selected_course_score
        final_recommended_courses = rec_df[['course_title', 'Similarity_Score', 'url', 'price', 'num_subscribers']]
        return final_recommended_courses.sort_values(by='Similarity_Score', ascending=False).reset_index(drop=True)
    
    # Handle keyword search if no exact match found
    keyword_matches = df[df['Clean_title'].str.contains(title_cleaned)]
    if not keyword_matches.empty:
        keyword_matches = keyword_matches[['course_title', 'url', 'price', 'num_subscribers']]
        keyword_matches = keyword_matches.drop_duplicates().reset_index(drop=True)
        return keyword_matches.head(numrec)

    # No match at all
    return f"No course or keyword match found for: '{title}'"


ans = recommend_course('Python', 10)
print(ans)
```
